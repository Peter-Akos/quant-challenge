```python
df = pd.read_csv('../data/united_data.csv', index_col=0).fillna(0)
y = df['Corn_Yield']
```

```python
y = pd.DataFrame(y)
y
```

```python
def convert_corn_yield(row):
    idx = row.name
    year = int(idx[:4])
    row['Corn_Yield'] = row['Corn_Yield'] + (1950-year)*1.96966271
    return row
    
```

```python
y = y.apply(convert_corn_yield, axis=1)
y
```
