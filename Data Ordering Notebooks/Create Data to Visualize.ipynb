```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/weather-0101-0930.csv', index_col=0)
```

```python
df
```

```python
cols = df.columns.to_list()
```

```python
avg_temps = cols[::4]
avg_temps
```

```python
mean_temperatures = df[avg_temps].mean(axis=1)
```

```python
mean_temperatures = pd.DataFrame(mean_temperatures, columns=['avg_temp'])
mean_temperatures
```

```python
mean_temperatures['index'] = mean_temperatures.index
```

```python
mean_temperatures['year'] = [np.nan]*len(mean_temperatures)
mean_temperatures['county'] = [np.nan]*len(mean_temperatures)
```

```python
mean_temperatures
```

```python
def get_year_and_county(row):
    idx = row['index']
    row['year'] = int(idx[:4])
    row['county'] = idx[5:]
    return row
```

```python
d = mean_temperatures.apply(get_year_and_county, axis=1)
```

```python
res = d.groupby('county')['avg_temp'].mean()
```

```python
res.to_csv('average_temperature_per_county.csv')
```
