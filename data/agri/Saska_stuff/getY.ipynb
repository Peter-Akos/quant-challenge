```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
county_df = pd.read_csv('../minnesota_county_yearly_agricultural_production.csv')
county_df
```

```python
county_df = county_df[['Year', 'County', 'Commodity', 'YIELD, MEASURED IN BU / ACRE']]
```

```python
county_df = county_df[county_df['Commodity'] == 'SOYBEANS']
```

```python
arr = []
for i in range(0, len(county_df)):
    arr.append(str(county_df.iloc[i]['Year']) + "-" + str(county_df.iloc[i]['County']).title() + " County")

arr
county_df["indexes"] = arr

df = county_df[['indexes', 'County', 'YIELD, MEASURED IN BU / ACRE']]
```

```python
df
```

```python
df = df.rename(columns={'indexes' : 'new_index', 'County' : 'County', 'YIELD, MEASURED IN BU / ACRE' : 'BU / ACRE'})
```

```python
df.to_csv('final_y_soybean.csv', index=False)
```
